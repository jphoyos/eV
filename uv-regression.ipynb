```python
import sklearn.model_selection
import sklearn.metrics
import json
import joblib

class Encoder(json.JSONEncoder):
    def default(self, obj): # pylint: disable=E0202
        if isinstance(obj, np.integer):
            return int(obj)
        elif isinstance(obj, np.floating):
            return float(obj)
        elif isinstance(obj, np.ndarray):
            return obj.tolist()
        elif isinstance(obj, sklearn.tree.DecisionTreeClassifier):
            return obj.__class__.__name__
        else:
            return super(Encoder, self).default(obj)
        
class Splitter:
    def __init__(self,data, label = "band_gap_mean"):
        self.data  = data
        self.features = self.data.iloc[:,:-2]
        self.label = self.data[label]
        self.data = None     
    def split(self, location):
        self.features_train, self.features_test, self.label_train, self.label_test = sklearn.model_selection.train_test_split(self.features, self.label, test_size=0.3 )
        self.features_train.to_csv(location + 'features_train.csv', index = False)
        self.features_test.to_csv(location + 'features_test.csv', index = False, header = True)
        self.label_train.to_csv(location + 'label_train.csv', index = False)
        self.label_test.to_csv(location + 'label_test.csv', index = False, header = True)
        
class EvModelDev:
    def __init__(self, features_train, label_train, models):
        self.features_train = pd.read_csv(features_train)
        self.label_train = pd.read_csv(label_train)
        self.train = np.concatenate([self.features_train, self.label_train], axis=1) #to speed up the training we convert it to numpy
        self.models = models
    def grid_search(self,report_location, n_jobs=4, models = models, scoring = 'geometric_mean' ):
        self.gs_scores = {}
        for model in models:
            nomdel_name = model[0].__class__.__name__
            estimator = model[0]
            grid = model[1]
            features = self.train[:,:-1]
            label = self.train[:,-1]
            if scoring == 'geometric_mean':
                scorer = geometric_mean_scorer
            grid_search = sklearn.model_selection.RandomizedSearchCV(
            estimator=estimator,
            param_distributions=grid,
            scoring=scorer,
            n_jobs=n_jobs,
            verbose=2,
            random_state=11,
            return_train_score=True,
            cv=3
            )
            grid_search.fit(features, label)
            self.gs_scores[model_name] = [grid_search.cv_results_, grid_search.best_params_, grid_search.best_score_]
            with open(report_location + 'grid_search_'+ model_name + '.json', 'w', encoding='utf-8') as json_file:
                     json.dump(self.gs_scores[model_name], json_file, indent=2, ensure_ascii=False, cls=Encoder)
    def train_direct(self, model_location):
        for estimator in self.models:
            model_name = estimator.__class__.__name__
            print(model_name)
            features = self.train[:,:-1]
            label  = self.train[:,-1]
            estimator.fit(features, label)
            print(estimator.score(features,label))
            joblib.dump(estimator, model_location + model_name + ".joblib")
    def predict_direct(self, models_location, features_test, label_test):
        features_test = pd.read_csv(features_test)
        label_test = pd.read_csv(label_test)
        for model in self.models:
            model_name = model.__class__.__name__
            estimator = joblib.load(models_location + model_name + '.joblib')
            try:
                bg_predict = estimator.predict(features_test.values)
            except:
                bg_predict = np.zeros(len(label_test),1)
            results = np.column_stack([label_test,bg_predict])
            np.savetxt(models_location + model_name + '_predictions.txt',results)

class EvalModel:
    def __init__(self, models):
        self.models = models
    def performance(self, models_location):
        for model in self.models:    #se debe crear un json que guarde todos los datos
            model_name = model.__class__.__name__
            results = np.loadtxt(models_location+ model_name + '_predictions.txt')
            mae = sklearn.metrics.mean_absolute_error(results[:,0],results[:,1].round(2))  #TWO decimal digits
            mse = sklearn.metrics.mean_squared_error(results[:,0],results[:,1].round(2))
            print('MAE: ', mae)
            print('\nMSE: ', mse)
```
